```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                          "bar", "bar", "bar", "bar"],
                    "B": ["one", "one", "one", "two", "two",
                          "one", "one", "two", "two"],
                    "C": ["small", "large", "large", "small",
                          "small", "large", "small", "small",
                          "large"],
                    "D": [1, 2, 2, 3, 3, 4, 5, 6, 7]})
```

```python
df
```

```python
from collections import defaultdict

df['E'] = df[['C', 'D']].apply(lambda x: {x[0]:[x[1]]}, axis=1)
```

```python
df
```

```python
sectors = df.groupby(['A', 'B'])
```

```python
sectors.groups.keys()
```

```python
sectors.get_group(('bar', 'one'))
```

```python
from functools import reduce

def merge(x, y):
    print(x)
    print(y)
    if isinstance(x, dict):
        for key in y.keys():
            if key in x.keys():
                x.update({key:x[key] + y[key]})
            else:
                x.update({key:y[key]})
    return x
    
    

pd.pivot_table(df, index=['A', 'B'], values=['E'], aggfunc=lambda x: reduce(merge, x))
```

```python
sectors.agg(lambda x: reduce(merge, x))
```
